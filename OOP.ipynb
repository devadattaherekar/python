```python
class Circle:
    PI=3.142 # class level object

    def __init__(self,radius=0):
        print("__init__ called")
        self.radius=radius
        
    def get_cirumference(self):
        return 2*self.PI*self.radius
    
    def get_area(self):
        return self.PI*self.radius*self.radius
    
    def __str__(self):
        print("__str__ called")
        return f"radius={self.radius}"

    def __repr__(self):
        print("__repr__ called")
        return f"<{self.radius}>"
    
shape1=Circle()
print(shape1.__repr__(),shape1.__str__())
shape2=Circle(3)
print(shape2)
print("PI is ",Circle.PI)
print("Cirumference is ",shape2.get_cirumference())
print("Cirumference is ",Circle.get_cirumference(shape2))
print("Area is ",shape2.get_area())
print("Area is ",Circle.get_area(shape2))
```

```python
class Employee:
    
    def __init__(self):
        self.employees=[]
        
    def accept_employee(self):
        self.name=input("Enter name ")
        self.sal=float(input("Enter sal "))
        record={"name":self.name,"sal":self.sal}
        self.employees.append(record)
    
    def __str__(self):
        return f'{self.employees}'
    
    def __repr__(self):
        return f'{self.employees}'
    
    def __getitem__(self,index):
        return f"{self.employees[index]}"
    
    
e1=Employee()
e1.accept_employee()
e1.accept_employee()
#e1.accept_employee()
print(e1)   
print("First record is ",e1.employees[0]) 
print("First record is ",e1[0]) 
```
