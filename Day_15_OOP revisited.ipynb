```python
class CustomIterable:
    def __init__(self):
        self.values=[10,20,30,40,50]
        self.index=0
        
    def __len__(self):
        return len(self.values)
    
    def __next__(self):
        if self.index < len(self.values):
            current=self.values[self.index]
            self.index+=1
            return current
        else:
            raise StopIteration
    
    def __iter__(self):
        return self
    

iter1=CustomIterable()
print(next(iter1))
print(next(iter1))
print(next(iter1))
print(next(iter1))
print(next(iter1))
#print(next(iter1))
iter2=CustomIterable()
print(sum(iter2))
iter3=CustomIterable()
for item in iter3:
    print(item,end=" ")
print(" ")
print("Total elements are ",len(iter3))
```

```python
class Complex:
    def __init__(self,real=0,imag=0):
        self.real=real
        self.imag=imag
    def __str__(self):
        return f"({self.real}+{self.imag}j)"
    def __repr__(self):
        return f"<{self.real} {self.imag}>"
    def __add__(self,cc):
        return Complex(self.real+cc.real,self.imag+cc.imag)
    def __sub__(self,cc):
        return Complex(self.real-cc.real,self.imag-cc.imag)
    def __mul__(self,cc):
        return Complex(self.real*cc.real,self.imag*cc.imag)
    def __truediv__(self,cc):
        print("__truediv__")
        return Complex(self.real/cc.real,self.imag/cc.imag)
    def __floordiv__(self,cc):
        print("__floordiv__")
        return Complex(self.real//cc.real,self.imag//cc.imag)
   
c1=Complex(10,11)
print(c1)
c2=Complex(3,5)
print(c2)
c3=c1+c2
print(c3)
c3=c1-c2
print(c3)
c3=c1*c2
print(c3)
c3=c1/c2
print(c3)
c3=c1//c2
print(c3)
```

```python
# < > >= <= == !=
# __lt__() __gt__() __ge__() __le__() __eq__() __ne__()

class Integer:
    def __init__(self,value=0):
        self.value=value
    def __str__(self):
        return f"{self.value}"
    def __lt__(self,ii):
        print("__lt__() called")
        return self.value < ii.value
    def __gt__(self,ii):
        print("__gt__() called")
        return self.value > ii.value
    def __ge__(self,ii):
        print("__ge__() called")
        return self.value >= ii.value
    def __le__(self,ii):
        print("__le__() called")
        return self.value <= ii.value
    def __eq__(self,ii):
        print("__eq__() called")
        return self.value == ii.value
    def __ne__(self,ii):
        print("__ne__() called")
        return self.value != ii.value
    def __pow__(self,ii):
        return self.value**ii.value
    
i1=Integer(2)
print(i1)
i2=Integer(15)
print(i2)
if (i1 == i2):
    print("Both are equal")
else:
    print("Not equal")
#print(i1**i2)
```

```python
class Book:
    def __init__(self,name):
        self.name=name
    def __str__(self):
        return f"You are reading {self.name} book"
    
class Library: # container class
    def __init__(self,*books): # contained object 
        self.books=books
    def __str__(self):
        for book in self.books:
            print(book)
        return f"Library contains {len(self.books)} books"
    
book1=Book("Head First Design Patterns")
#print(book1)
book2=Book("Effective C++ Scott Meyers")
#print(book2)
book3=Book("101 ways to learn python")
#print(book3)

lib1=Library(book1,book2,book3)
print(lib1)
```

```python
class Employee:
    """
    Employee class with id name company
    """
    def __init__(self,emp_id,emp_name,company):
        self.emp_id=emp_id
        self.emp_name=emp_name
        self.company=company
        
    @property
    def get_company(self): # accessor functions
        return self.company
    
    @get_company.setter
    def set_company(self,company):
        companies=["Google","Amazon","Microsoft","TomTom"]
        if company in companies:
            self.company=company
        else:
            raise ValueError(f"You are not allowed to work for {company}!")
            
    def __str__(self):
        return f"{self.emp_name} with id {self.emp_id} works for {self.company}"
    
    @staticmethod
    def static_method(): # static method can never have self parameter!
        print("This is static method")
        
    @classmethod
    def class_method(cls):
        print("This is class method and class name is ",cls.__name__)
        print(Employee.__doc__)
        
        
        
e1=Employee(12345,"Rajesh","TomTom")
print(e1)
#print("Company is ",e1.get_company())
print("Company is ",e1.get_company)
#e1.set_company("Google")
e1.set_company="Google"
print(e1) 
Employee.static_method() # Note: use class name to call static methods
# static methods belong to class not instances 
if Employee.__name__=="Employee":
    Employee.class_method()
```

```python
# abstract class cannot be instantiated (you cannot create object of that class)
# a class which is inherited from abstract class also abstract by default
# You have to override abstract method in derived class so that it becomes concreate class
# Your class should be derived from ABC class and use abstractmethod decorator on abstract method
from abc import ABC,abstractmethod
class Vehicle(ABC): # Here Vehicle class is abstract class as it contains abstract methods
    def __init__(self,name):
        self.name=name
        
    @abstractmethod
    def start(self): # abstract methods do not have any implementation 
        pass
    
    @abstractmethod
    def stop(self):
        pass
    
    def carry(self): # Abstact class also contains non abstract methods
        print("Vehicle carries goods")
         
    def __str__(self):
        return f"{self.name}"
   
class Chargeable(ABC): # chargeable interface
    @abstractmethod
    def charge(self):
        print("Charging...please wait!")
        
class Fourwheelers(Vehicle):
    def __init__(self,name):
        super().__init__(name)
    def start(self):
        print(f"{super().__str__()} starts with full speed")
    def stop(self):
        print(f"{super().__str__()} stops")
    
class Truck(Vehicle):
    def __init__(self,name):
        super().__init__(name)
    def start(self):
        print(f"{super().__str__()} starts with normal speed")
    def stop(self):
        print(f"{super().__str__()} truck stops")

        
class Tata_Truck(Truck,Chargeable):
    def __init__(self,name):
        super().__init__(name)
    def start(self):
        print(f"{super().__str__()} starts with normal as well as full speed")
    def stop(self):
        print(f"{super().__str__()} tata truck stops")
    def charge(self):
        super().charge()
        print(f"{super().__str__()} is charging!")

#car=Vehicle("audi")  # one cannot create objects of abstract class
car=Fourwheelers("mercedez")
car.start()
car.carry()
car.stop()
truck=Truck("volvo")
truck.start()
truck.carry()
truck.stop()
tata=Tata_Truck("1210 truck")
tata.charge()
```

```python
from abc import ABC,abstractmethod

class Warrior(ABC):
    def __init__(self,name):
        self.name=name
    @property
    def get_name(self):
        return self.name
    @abstractmethod
    def attack(self):
        pass
    
class Swordsman(Warrior):
    def __init__(self,name):
        super().__init__(name)
    def attack(self):
        print(f"swordsman {super().get_name} attacks with sword")

class Crossbowman(Warrior):
    def __init__(self,name):
        super().__init__(name)
    def attack(self):
        print(f"crossbowman {super().get_name} attacks with bow and arrow!")
        
class Dragon(Warrior):
    def __init__(self,name):
        super().__init__(name)
    def attack(self):
        print(f"Dragon {super().get_name} attacks with fire!")
    def fly(self):
        print(f"Dragon {super().get_name} flies high in the sky...")
        
hero1=Swordsman("Willam Wallace")
#hero1.attack()
hero2=Crossbowman("Achilles")
#hero2.attack()
hero3=Dragon("Drago")
#hero3.attack()
#hero3.fly()
for fighter in [hero1,hero2,hero3]:
    if type(fighter) is Swordsman:
        fighter.attack()
    elif type(fighter) is Crossbowman:
        fighter.attack()
    elif type(fighter) is Dragon:
        fighter.attack()
        fighter.fly()
        
def fight(fighter):
    fighter.attack()
fight(hero1)
fight(hero2)
fight(hero3)
```
