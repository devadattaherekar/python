```python
import numpy as np

list1=[10,20,30,40,50]
print(type(list1),list1)

nparr=np.array([10,20,30,40,50])
print(type(nparr),nparr)


# np.array is faster than list, You can perform some computation and store val in list
# 1.append()
list2=[]
for val in list1:
    list2.append(val*100)
print(list2)

# 2.comprehension
list2=[val*100 for val in list1]
print(list2)

# 3. map
def fun(val):
    return val*100
for val in map(fun,list1):
    print(val,end=" ")
print(" ")
print(list(map(fun,list1)))
print(list(map(lambda val:val*100,list1)))

# 4. numpy uses ufuncs (universal functions) and performance is fast

nparr=np.array([3,4,5]) # It can only store same dtypes
print(nparr.dtype)
print(nparr*100)
nparr=np.array([3,4.0,5])# implicit upcasting
print(nparr.dtype)
```

```python
# indexing using callables

import pandas as pd
from string import ascii_lowercase
from string import ascii_uppercase

label_alpha=pd.Series(data=list(ascii_lowercase),index=list(ascii_uppercase))

print(label_alpha.loc[lambda x:[True for x in range(x.size)]])

def every_fifth(item):
    return [True if item%5==0 else False for item in range(item.size)]

print(label_alpha.iloc[lambda item:[True for item in range(item.size)]])
print(label_alpha.iloc[every_fifth])
```

```python
# selection by label or selection by position
# 1. series['label'] # slices, callables, boolean masks
# 2. series.loc['label'] # slices, callables, boolean masks
# 2. use series.iloc[position] 
# 3. series.label  # no slice or boolean mask support
# 4. series.get('label') # no slicing but provides default 

import pandas as pd

cities=['mumbai','bengaluru','chennai']
st_series=pd.Series(data=cities,index=['maharashtra','karnataka','tamilnadu'])
print(st_series.loc['maharashtra':'tamilnadu']) 
print(st_series['maharashtra':'tamilnadu'])
print(st_series.get('maharashtra')) # get returns nothing if label is invalid
# get works like loc i.e. labeled indexing and iloc i.e. integer indexing both!
print(st_series.get(1)) # returns bengaluru
print(st_series.get('delhi',default="cannot find value with that label"))
# Here default parameter can take int or dict as values# selection by label or selection by position
```

```python
# 1. create a series of length 100 containing squares of ints from 0 to 99 and assign it to variable squares
# 2. Extract the last 3 items from the squares using square bracket indexing
# 3. Repeat the above step using tail() method
# 4. Verify o/p of step2 and step3 is same using equals() method

import pandas as pd

squares=pd.Series(data=[item*item for item in range(100)])
print(squares)
print(squares[-3:])
print(squares.iloc[-3:])
print(squares.get(-3))
print(squares.tail(n=3))
print(squares.tail(n=3).equals(squares[-3:]))
```
